定义函数为 $y=2x+1$

```python
x_data = [2.0, 3.0, 4.0]
y_data = [5.0, 7.0, 9.0]
```

```python
W = np.arange(0.0, 4.0, 0.1)
b = np.arange(0.0, 4.0, 0.1)
[w, b] = np.meshgrid(W, b)
[w, b]
```

```python
mse_list = []
l_sum = 0

for x_val, y_val in zip(x_data, y_data):
  y_pred_val = forward(x_val)
  loss_val = loss(x_val, y_val)
  l_sum += loss_val
  mse = l_sum / 3
  mse_list.append(mse)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection='3d')
ax.plot_surface(w, b, mse)

plt.show()
```

`[X,Y] = meshgrid(x,y)` 将向量x和y定义的区域转换成矩阵X和Y,其中矩阵X的行向量是向量x的简单复制，而矩阵Y的列向量是向量y的简单复制

np.meshgrid() 函数主要用来绘制图形
