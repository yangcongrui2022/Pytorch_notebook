```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset, DataLoader
```

Dataset 和 DataLoader 一般结合使用

Dataset 为抽象类，不能实例化对象，主要用于构造数据集

DataLoader 需要获取 DataSet 提供的索引 [i] 和 len;用来帮助我们加载数据，比如做 shuffle(提高数据集的随机性)，batch_size，能拿出 Mini-Batch 进行训练。能帮我们自动完成这些工作。


```python
class DiabetesDataset(Dataset):
  def __init__(self, filepath):
    data = np.loadtxt(filepath, delimiter=',', dtype=np.float32)
    self.len = data.shape[0]
    self.x_data = torch.from_numpy(data[:, :-1])
    self.y_data = torch.from_numpy(data[:, [-1]])
  def __getitem__(self, index): # __getitem__目的是为支持下标(索引)操作，必须被重写
    return self.x_data[index], self.y_data[index]
  def __len__(self): # __len__目的是返回数据集大小，可以不重写
    return self.len

dataset = DiabetesDataset('diabetes.csv.gz') # 实例化DiabetesDataset
trainLoader = DataLoader(dataset=dataset, # 实例化DataLoader
              batch_size=32,
              shuffle=True,
              num_workers=0)
```

```python
class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear1 = torch.nn.Linear(8, 6)
    self.linear2 = torch.nn.Linear(6, 4)
    self.linear3 = torch.nn.Linear(4, 1)

    self.sigmoid = torch.nn.Sigmoid()

  def forward(self, x):
    x = self.sigmoid(self.linear1(x))
    x = self.sigmoid(self.linear2(x))
    x = self.sigmoid(self.linear3(x))
    return x

model = Model()
```

```python
criterion = torch.nn.BCELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

epoch_list = []
loss_list = []
```

```python
if __name__ == '__main__':
  for epoch in range(1000):
    for i, (inputs, labels) in enumerate(trainLoader, 0):
      y_pred = model(inputs)
      loss = criterion(y_pred, labels)

      epoch_list.append(epoch)
      loss_list.append(loss.item())

      optimizer.zero_grad()

      loss.backward()
      optimizer.step()

      if epoch % 100 == 0:
        print("epoch=", epoch, "loss=", loss.item())
```
